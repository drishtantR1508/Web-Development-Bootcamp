```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fact(n):
    b = 1
    for i in range(1,n+1,1):
        b = b * i
    return b
```

```python
x = np.arange(1,8,1)
y1 = np.array([fact(i-1) for i in x])
```

```python
print(x)
print(y1)
```

```python
plt.figure(figsize=(20,5))
plt.plot(x, y1,"b*-")
plt.xlabel('n/n-1')
plt.ylabel("(O(n-1)!)")
plt.title("Naive approach!")
```

```python
def n22n(n):
    return (n**2)*(2**n)
```

```python
x = np.arange(1,8,1)
y1 = np.array([n22n(i) for i in x])
```

```python
plt.figure(figsize=(25,10))
plt.plot(x, y1,"b*-")
plt.xlabel('n')
plt.ylabel("O(n^2 * 2^n)")
plt.title("DP Solution!")
```
